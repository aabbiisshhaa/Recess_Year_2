# KMeans Method

## 1. Load the dataset

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the data
df = pd.read_csv('customer_data_records.csv')

# Display first few rows of the dataframe
df.head()
```

## 2. Extract relevant features for clustering

```python

features = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(features)
```

```python
df['Cluster'] = kmeans.labels_

print(df)
```

## 3. Cluster Chart 1: Annual Income vs Spending Score

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segments based on Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
    
```
